### Data Manipulation


Filling the Null values with the appropriate values using aggregate functions such as Mean,Median,Mode

```python
# Defining the function to fill the null values with the median.
def data_median(series):
    return series.fillna(series.median())
```

```python
play_store_data.Rating = play_store_data['Rating'].transform(data_median)
```

```python
# Count the number of values in each column
play_store_data.isnull().sum()
```

Now, we have treated Rating  with median values


where some categorical variables such as "Type","Current Ver","Android Ver"  also have null values we treat it with mode values



```python
# Mode of categorical values
print(play_store_data['Type'].mode())
print(play_store_data['Current Ver'].mode())
print(play_store_data['Android Ver'].mode())
```

```python
# Filling the missing categorical values with Mode
play_store_data['Type'].fillna(str(play_store_data['Type'].mode().values[0]),inplace=True)
play_store_data['Current Ver'].fillna(str(play_store_data['Current Ver'].mode().values[0]),inplace=True)
play_store_data['Android Ver'].fillna(str(play_store_data['Android Ver'].mode().values[0]),inplace=True)
```

```python
# Counting the number of null values in each column
play_store_data.isnull().sum()
```

```python
play_store_data.head(7)
```

```python
# Converting Price,Size and installs

#we create a function which convert Mega-bite to kilo-byte

def convrt(Size):
    if Size =='Varies with device':
        return np.nan

    elif 'M' in Size:
        Size=Size.replace('M','000')
        return eval(Size)

    elif 'k' in Size:
        Size=Size.replace('k','')
        return eval(Size)
```

```python
#applying the convte function to convert in into kilo byte
play_store_data['Size']= play_store_data['Size'].apply(convrt)
play_store_data.head()
```

```python
# To remove the '+' and ',' from the number of install

# we make a function and apply it to the dataset

def removing_plus_comma(installs):
    if isinstance(installs, float):
        return installs
    elif '+' in installs:
        installs = installs.replace('+','')
    if ',' in installs:
        installs = installs.replace(',','')
    return installs
```

```python
play_store_data['Installs']= play_store_data['Installs'].apply(removing_plus_comma)
play_store_data['Installs']=play_store_data['Installs'].astype(float)
play_store_data.head()
```

```python
#converting $ to " " and make it in float
play_store_data['Price'] = play_store_data['Price'].str.replace('$', '').str.strip()
play_store_data['Price'] = play_store_data['Price'].astype(float)
```

```python
# making Last Updated in proper format
play_store_data['Last Updated']

#changing date format of 'Last Updated'
play_store_data['Last Updated'] = pd.to_datetime(play_store_data['Last Updated'])
```

# 6. What are the Top 10 installed apps in any category?

```python
def findtop10incategory(str):
    str = str.upper()
    top10 = play_store_data[play_store_data['Category'] == str]
    top10apps = top10.sort_values(by='Installs', ascending=False).head(10)
    plt.figure(figsize=(15,12))
    plt.title('Top 10 Installed Apps',size = 20);
    graph = sns.barplot(x = top10apps.App, y = top10apps.Installs)
    graph.set_xticklabels(graph.get_xticklabels(), rotation= 45, horizontalalignment='right');
```

```python
findtop10incategory('COMMUNICATION')
```
